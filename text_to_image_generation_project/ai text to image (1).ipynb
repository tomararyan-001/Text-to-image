```python

!pip install flask pyngrok diffusers torch transformers
from flask import Flask, request, send_file
from pyngrok import ngrok
from diffusers import StableDiffusionPipeline
import torch
from io import BytesIO
!ngrok authtoken 'enter  your ngrok auth token here' #login on ngrok website and generate auth-token then replace it.
import os
os.makedirs("static/generated_images", exist_ok=True)
#Initialize the Stable Diffusion model
sd_model = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4").to("cuda")
#set up Flask application
app = Flask(__name__)
@app.route("/generate-image", methods=["POST"])
def generate_image():
  data = request.get_json()

  prompt = data.get("text", "")

  image = sd_model(prompt).images[0]

  img_io = BytesIO()

  image.save(img_io, "PNG")

  img_io.seek(0)

  return send_file(img_io, mimetype="image/png")

public_url = ngrok.connect(8000)

print("Public URL:", public_url)



if __name__ == "__main__":

  app.run(port=8000)
```
